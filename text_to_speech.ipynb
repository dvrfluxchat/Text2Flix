```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio
from pathlib import Path

# download and load all models
preload_models()
```

```python

text_prompt = """
Coorg mein aapka swagat hai, ek aakarshak pahadi sthal hai jo Karnataka, Bharat mein sthit Western Ghats mein basa hua hai. "India ka Scotland" ke roop mein mashhoor Coorg, apne tasveeri nazare, sugandhit coffee ke bagicha, dhundhle pahadon aur samriddh sanskritik virasat ke liye mashhoor hai.
"""
# generate audio from text

history_prompt = "v2/hi_speaker_0"
audio_array = generate_audio(text_prompt, history_prompt=history_prompt)
file_name = f"coorg_{Path(history_prompt).name}.wav"
# save audio to disk
write_wav(file_name, SAMPLE_RATE, audio_array)
  
# # play text in notebook
Audio(audio_array, rate=SAMPLE_RATE)
# List available 🐸TTS models and choose the first one
# model_name = TTS.list_models()[0]
# # Init TTS
# tts = TTS(model_name)
```
